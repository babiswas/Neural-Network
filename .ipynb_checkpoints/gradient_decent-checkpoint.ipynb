```python
def apply_transformation(func,*args):
    result=list()
    for i in args:
        result.append(func(i))
    return result

def func(i):
    return 2*i

print(apply_transformation(func,1,2,3,4))
```

```python
def diff(func,x):
    y=func(x)
    delta=0.000000000001
    y1=func(x+delta)
    return (y1-y)/delta

def func_sq(x):
    return x**2

diff(func_sq,2)
```

```python
def partial_derv(func,index,*args):
    y=func(*args)
    delta=0.000000001
    args=list(args)
    print(args)
    args[index]+=delta
    y1=func(*args)
    return (y1-y)/delta

def func(x1,x2,x3):
    return 2*x1+2*x2+x3

partial_derv(func,1,1,2,3)
```

```python
import numpy as np
def gradient_dscnt(X,Y):
    m=b=0
    iteration=1400
    n=len(X)
    learning_rate=0.0001
    for i in range(iteration):
        Y_pred=m*X+b
        print(2*sum((Y-Y_pred)**2)/n)
        dm=-(2/n)*sum(X*(Y-Y_pred))
        db=-(2/n)*sum(-(Y-Y_pred))
        m=m-learning_rate*dm
        b=b-learning_rate*db
gradient_dscnt(np.array([1,2,3,4]),np.array([10,12,5,6]))
```
